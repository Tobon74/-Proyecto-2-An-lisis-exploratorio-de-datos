```python
# Reimportar librerías después del reinicio
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Recargar el archivo Excel original
file_path = "Proyecto.xlsx"  # Corrected file path
df = pd.read_excel(file_path, sheet_name='Life Expectancy Data')
```

```python
# Asegurar limpieza de columnas
df.columns = df.columns.str.strip()
```

```python
# Selección de columnas y limpieza de valores faltantes
relevant_cols = ["Alcohol", "BMI", "Adult Mortality", "Schooling", "GDP", "Life expectancy"]
df_cleaned = df[relevant_cols].dropna()
```

```python
# Separar variables predictoras (X) y objetivo (y)
X = df_cleaned[["Alcohol", "BMI", "Adult Mortality", "Schooling", "GDP"]]
y = df_cleaned["Life expectancy"]
```

```python
# Ajustar modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)
```

```python
# Obtener coeficientes e intercepto
regresion_df = pd.DataFrame({
    "Variable": X.columns,
    "Coeficiente": model.coef_
})  # <- Aquí cerramos correctamente el paréntesis
```

```python
# Agregar el intercepto como una fila adicional
regresion_df.loc[len(regresion_df.index)] = ["Intercepto", model.intercept_]
```

```python
# Simular un caso de predicción
nuevos_datos = pd.DataFrame({
    "Alcohol": [6.2],
    "BMI": [23.5],
    "Adult Mortality": [110],
    "Schooling": [13.2],
    "GDP": [9000]
})
```

```python
# Hacer predicción
prediccion = model.predict(nuevos_datos)
```

```python
# Crear tabla con la predicción
pronostico_df = nuevos_datos.copy()
pronostico_df["Life expectancy estimada"] = prediccion[0]
```

```python
# Mostrar los resultados en pantalla (opcional pero útil)
print("📊 Coeficientes del modelo:")
print(regresion_df)
print("\n🔮 Pronóstico de expectativa de vida para el nuevo caso:")
print(pronostico_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma de la expectativa de vida
plt.figure(figsize=(10, 6))
sns.histplot(df["Life expectancy"].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Distribución de la Expectativa de Vida")
plt.xlabel("Life Expectancy (años)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
# Promedio de expectativa de vida por país (top 10 países con mayor promedio)
top10 = df.groupby("Unnamed: 0")["Life expectancy"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top10.values, y=top10.index, palette="viridis")
plt.title("Top 10 países con mayor expectativa de vida promedio")
plt.xlabel("Life Expectancy Promedio")
plt.ylabel("País")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo gráfico
sns.set(style="whitegrid")

# Gráfica de barras: expectativa de vida promedio por tipo de país (Status)
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="Status", y="Life expectancy", estimator="mean", ci=None, palette="Set2")
plt.title("Expectativa de Vida Promedio según Nivel de Desarrollo")
plt.xlabel("Tipo de País")
plt.ylabel("Life Expectancy Promedio")
plt.tight_layout()
plt.savefig("Barplot_Por_Estado.png")
plt.show()
```
